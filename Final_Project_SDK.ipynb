```python
#Part 2 of Final Project
#We are going to Calculate the Mean, Media, and Mode for Each Columns That Makes Statistical Sense

import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/EtherVerseCodeMate/Python4DataAnalytics/main/New_York_City_Community_Health_Survey.csv')


# Calculating statistics for each numerical column
for col in df.select_dtypes(include=[np.number]).columns:
    print(f"\nColumn: {col}")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    try:
        print(f"Mode: {df[col].mode()[0]}")  # mode is a Series, so we get the first value
    except IndexError:  # catches cases where mode() returns an empty Series (no mode)
        print("Mode: No mode found")
    print(f"Standard Deviation: {df[col].std()}")

```

```python
#Part 3 of Final Project

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv('https://raw.githubusercontent.com/EtherVerseCodeMate/Python4DataAnalytics/main/New_York_City_Community_Health_Survey.csv')

# Part 1: Creating distribution plots
# Choose the columns you're interested in
columns_to_plot = ['No Health Insurance', 'Smoking Status (current smokers)', 'Binge Drinking', 'Obesity', 'Flu shot in last 12 months, adults ages 65+ (not age-adjusted)','Self-reported Health Status (excellent/very good/good)']  # Replace with your actual column names

for column in columns_to_plot:
    plt.figure(figsize=(10,5))

    # Histogram
    plt.subplot(1,2,1)
    df[column].plot(kind='hist', edgecolor='black')
    plt.title(f'Histogram of {"No Health Insurance"}')

    # Boxplot
    plt.subplot(1,2,2)
    df[column].plot(kind='box')
    plt.title(f'Boxplot of {"Self-reported Health Status (excellent/very good/good)"}')

    plt.tight_layout()
    plt.show()

# Part 2: Calculating statistical measures
for column in df.columns:
    if np.issubdtype(df[column].dtype, np.number):
        print(f'{column}:\n Mean: {df[column].mean()} \n Median: {df[column].median()} \n Mode: {df[column].mode()[0]} \n Std Dev: {df[column].std()}\n')

# Part 3: Creating correlation graphs
# First, calculate the correlation matrix
correlation_matrix = df.corr()

# Then, plot the correlations for the top 5 correlations
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Find top 5 correlated pairs
corr_pairs = correlation_matrix.unstack()
sorted_pairs = corr_pairs.sort_values(key=abs)
top_pairs = sorted_pairs[-6:-1]  # We ignore the top 1 as it is the correlation of a variable with itself

print(top_pairs)
```
